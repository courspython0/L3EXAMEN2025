```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor


df = pd.read_csv(r'C:/examPy_Defense/bdd/bddBostonNettoyer_sans_aberrante.csv')


print(df.head())
print(df.head())
print(df.info())
print(df.describe())





PRIX_DU_LOGEMENT = df['PRIX_DU_LOGEMENT']
x = df[[
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'CONCENTRATION_NOX',
    'NBR_MOYEN_PIECES',
    'ANCIENNETE_LOGEMENT',
    'DISTENCE_ENTREPRISES',
    'ACCES_AUTOROUTE',
    'TAUX_TAXE_FONCIERE',
    'RATION_ELEVE_ENSEIGNANT',
    'DEMOGRAPHIE_MINORITAIRE',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
]]
```

```python

"""
    Notre dataset separé en deux la variable cible qui est (PRIX_DU_LOGEMENT)
    et nos varibales clé qui sont dans la liste x
    seront séparé en jeu de test et d'entrainment 

    l'objectif ici est d'obtenir une bonne base pour l'entrainement 
    car il est important de noté que la performance de notre modèle y dépedend 
    voila la raison por laquelle on a prit 80% de notre data set pour la dédié 
    au jeu d'entrainement , contre seulement 20% pour le test 

    il est important de noter que cette partie est cruciale pour 
    notre projet car c'est d'elle qu'on  saura si notre modele fait du overfitting 
    ou soit de l'underfitting 

"""



X_train,X_test,y_train,y_test = train_test_split(x,PRIX_DU_LOGEMENT,test_size=0.3, random_state=0)

```

```python

"""
    lorsqu'on utilise : 
    les algorithme comme :  lasso(L1), Ridge(L2), régression linaire il est important de 
    normaliser pour éviter que le model concidère
    une features plus que les autre ( c'est comme mettre toutes les variable au meme niveau
    de comparaison )
    la normalisation permet a ridge ,lasso et régression linaire de se baser sur l'influences de chaqes variable 
    d'entre sur la prédiction au lieu de ce concentre uniquement sur leurs valeurs numérque 
"""


scaler = StandardScaler()
X_train_Normaliser = scaler.fit_transform(X_train)
X_test_Normaliser = scaler.transform(X_test)

```

```python


"""
    On instancie notre algorithme linéaire  RandomForestRegressor()
    pour prepare, entrainer et tester notre model

    Note pour culture général RF veut dire foret de aléatoire

"""

rf_regressor = RandomForestRegressor(n_estimators=100, max_depth=20,random_state=0)

"""Entrainement de notre model avec la fonction fit()"""
rf_regressor.fit(X_train,y_train)

"""
    Ont test notre modele avec les données dont il n'a jamais 
    lors d'entrainement pour voir si on est dans le underfitting ou soit de l'overfitting 


"""
```

```python

y_predis = rf_regressor.predict(X_test)


"""
    On essaie de voir la correlation qui exite entre les prix des logement predit
    et les vrai prix du logement présant dans le dataset 

"""


plt.scatter(y_predis,y_test)
plt.show()

```

```python



"""Les métrique d'erreurs"""


MAE = mean_absolute_error(y_test,y_predis)
MES = mean_squared_error(y_test,y_predis)
"""R2 Coefficient de détermination , plus se proche de 1 , cela indiqu notre modèle est bon"""
r2 = r2_score(y_test,y_predis)
RMES = np.sqrt(MES)

print(f'mean_absolute_error {MAE}')
print(f'mean_squared_error {MES}')
print(f'Coefficient de détermination R2 {r2}')
print(f'la racine carre du mean_squared_error {RMES}')


```

```python





"""
    LA COURBE DE RESIDUS(ERREURS)
    ce graphique nous permet de voir la distribution (répartition d'erreurs)
    d'erreurs entre le valeurs prédis et le vrai valeurs observé dans l'univers réel
    erreurs = y(reel) - y(predis)
    notes : 
        une distribution aleatoire : est un bon 
        signe cela nous montre que notre modèle n'est se trompe 
        pas assez (faible taux d'ereurs), les erreurs sont bien repartie

        tendance(ou des motif ) : cela veut dire que 
        notre modèle commet des érreurs 

"""

"""calcul des résidus"""

residus = y_test - y_predis

plt.figure(figsize=(10,6))
sns.residplot(x = y_predis, y = residus,color='g')
plt.axhline(y = 0, color = 'r', linestyle = '--')

plt.title('COURBE DE RESIDUS')
plt.xlabel('Prix des logement prédis')
plt.ylabel('residus')
plt.show()

```

```python


"""

    Note générale


    le random forest utilise deux type d'algorithme 
    la regression et la calssification 

    l'objectif du RF est de divisiser un dataframe(ou dataset ou encore une BDD) en plusieurs petit 
    échantillons(appelé arbre de decision ) avec possibilité de répetition  dans le but de trouver different reponse et 
    les reponses obtenus sont en suite votés pour le cas de la calssification et pour le 
    cas de la regression on fait la moyenne de reponse(resultat obtenus) pour estimé la reponse finale

    les étape clé : 
    1) avoir un model(dataframe ou dataset)
    2) faire le nétoyage du dataframe ou dataset
    3) on creer un algorithme RF(soit la calssification ou soit la regression) 
    4) on separe nos donnée une partie pour le test et l'autre pour l'entrainement
    5) on détéermine le nombre des arbres qu'on veut cree partant de notre DF avec 
        n_estimators=100, 
    6) on détérmine le nombre de profondeurs de chaque arbre(càd le nombre total de question 
        que l'algorithme va se posé pour chaque arbre avec 
        max_depth=20
    7) on entraine notre model avec la fonction fit 
    8) on fait les prédiction avec la fonction predics

    9) testé la perforcemence de notre model avec 
        MSE "plus se proche du zeros mieux c'est"
        R2 "plus se proche de 1 mieux c'est"


    le overfitting c'est quand un model memorise le données d'entrainement au lieu d'apprendre les 
    tendance générales  

    l'overfitting arrive quand :

    1)le données d'entrainement contiennent bcp d'erreurs (bruits)
    (les valeurs null, des types incoherent dans une meme variable(feature))
    2) les données d'entrainement est trop petit 
    3) complexité du model (pour le cas par exemples du random forest regressor ) on a un 
    une bdd qui contient bcp des échantillons (arbre de decision) avec un profondeurs de 1000
    par exemples par échantillons


    le underfitting : ici le underfitting n'arrive meme pas à mémoriser car le quantité 
    de donnée d'entrainement est tellement faible (un model simple)
    arrive quand le données d'entrainement est simple c'est qui empeches le model qui 
    sera mise en place à etre performant au vrai valeurs des test 

    alors c'est quoi le vrai model?
    un vrai model c'est un model qui au moment de teste la MSE trouvé est égale au 
    MSE de l'entrainement

    si le MSE de teste est plus grand que le MSE d'entrainement 
    c'est qui a un overfitting(surajustement) le model n'a pas généralesé mais 
    a peut-etre bloqué (suite au cause agncés ci-haut)


  
    l'objectif est de trouver un MSE de teste plus bas par rapport au MSE de entrainement
    ce qui prouverait que notre model lorsqu'il travaille avce de vrai valeurs ils ne se trompe pas






"""
```
