```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler


"""

    ici nous recupérons notre dataset (dataframe)
    dans son état brut telque téléchargé depuis le 
    lien donné par le professeur

"""


df = pd.read_csv(r"HousingData.csv")



"""
    Nous essayons de voir un peu à quoi ressemble notre dataset telque téléchargé dans 
    son état brut avec la fonction info()et describe()
"""
print('--VISUALISATION DU DATASET--')

print(df.info())
print(df.describe())
```

```python

"""
    nettoyage du model HousingData.csv
    c'est à dire notre dataframe df.
    Nous renomons les colonnes du dataframe 
    pour une meilleures gestion et une bonne comprehension 
    chaque colonne est renommées en francais 

"""

print("""---------
        nettoyage du model HousingData.csv\n
        c'est à dire notre dataframe df.\n
        Nous renomons les colonnes du dataframe \n
        pour une meilleures gestion et une bonne comprehension\n 
        chaque colonne est renommées en francais 

    -------------"""
      )



df.rename(columns =
          {
              'CRIM' : 'TAUX_DE_CRIMINALITE',
              'ZN': 'zonage_residentiel',
              'INDUS': 'PROPORTION_INDUSTRIELLE',
              'CHAS' : 'BORD_DU_FLEUVE',
              'NOX' : 'CONCENTRATION_NOX',
              'RM' : 'NBR_MOYEN_PIECES',
              'AGE' : 'ANCIENNETE_LOGEMENT',
              'DIS' : 'DISTENCE_ENTREPRISES',
              'RAD' : 'ACCES_AUTOROUTE',
              'TAX' : 'TAUX_TAXE_FONCIERE',
              'PTRATIO' : 'RATION_ELEVE_ENSEIGNANT',
              'B' : 'DEMOGRAPHIE_MINORITAIRE',
              'LSTAT': 'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
              'MEDV': 'PRIX_DU_LOGEMENT'

            }, inplace = True
)

"""

    On visualise notre dataset après les modifications apportées
    Note : la fonction .head() n'affiche que 5 lignes par défaut 
    si on ne precise pas le nombre de ligne dans le parametre 

"""

print(  """ 
            On visualise notre dataset après les modifications apportées
            Note : la fonction .head() n'affiche que 5 lignes par défaut 
            si on ne precise pas le nombre de ligne dans le parametre
        """
    )
print(df.head())

```

```python

"""
    On modifier les types de ces differentes colonnes 
    à float pour l'unicté de type lorsqu'on aura 
    à travailler avec notre dataset
    
"""

df.TAUX_TAXE_FONCIERE = df.TAUX_TAXE_FONCIERE.astype(float)
df.ACCES_AUTOROUTE = df.ACCES_AUTOROUTE.astype(float)
df.RATION_ELEVE_ENSEIGNANT = df.RATION_ELEVE_ENSEIGNANT.astype(float)



"""
    On supprime les valeurs null des ces differentes colonnes
    pour n'est pas avoir de l'overfitting dans la suite de notre 
    travaille avec la fonction dropna()

"""

df.dropna(subset = [
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'BORD_DU_FLEUVE',
    'ANCIENNETE_LOGEMENT',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR'], 
    inplace=True
)





"""

    pour ne pas géré les valeurs null présentes dans chaque colonnes 
    on utilise la fonction dropna sur l'ensembe de notre dataset
    Note : inplace= True ; c'est pour que ces modifications soit effectif
    directement  dans notre dataset


"""


df.dropna(inplace= True)

print('-------VISUALISATION DU DATASET SANS VALEURS NULLS--------')
print(df.info())
print(df.describe())




```

```python

print(df['BORD_DU_FLEUVE'].unique)
"""
    on supprime la variable BORD_DU_FLEUVE car cette feature(variable) 
    ne contient qu'une seule  valeure qui est zero(0)
    or notre model d'apprentissage son objectif est de s'entrainer 
    avec les valeurs differente 
    pour bien esptimer(predire) le pris des logement 
    c'est pour cette raison que nous supprimons le feature BORD_DU_FLEUVE
    une facons d'éviter le overfitting

    Note : axis=1; pour signaler à python qu'il s'agit bien de la suppression 
    d'une colonne et non d'une ligne axis=0 
"""
df.drop('BORD_DU_FLEUVE', inplace=True, axis=1)

"""

    Visualisation du datset apres suppression du features BORD_DU_FLEUVE

"""

print('--Visualisation du datset apres suppression du features BORD_DU_FLEUVE--')
print(df.head())
print(df.info())
print(df.describe())





```

```python


"""
    On enregistre toute les modfication apportées à notre dataset HousingData.csv 
    dans un fichier excel et csv
"""
df.to_excel(r'C:\examPy_Defense/bdd/bddBostonNettoyer.xlsx')
df.to_csv(r'C:\examPy_Defense/bdd/bddBostonNettoyer.csv')

""""
    Dans le cadre de ce projet notre seule et unique feature(variale)
    cible est belle est bien la vaiable PRIX_DU_LOGEMENT
    car il s'agit de détérminer la variation du  PRIX de LOGEMENT à BOSTON 
    en faction des critères x et z,w etc... qui sont pour nous 
    les variables clés

"""



PRIX_DU_LOGEMENT = df['PRIX_DU_LOGEMENT']
x = df[[
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'CONCENTRATION_NOX',
    'NBR_MOYEN_PIECES',
    'ANCIENNETE_LOGEMENT',
    'DISTENCE_ENTREPRISES',
    'ACCES_AUTOROUTE',
    'TAUX_TAXE_FONCIERE',
    'RATION_ELEVE_ENSEIGNANT',
    'DEMOGRAPHIE_MINORITAIRE',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
]]

```

```python


"""

Notre dataset separé en deux la variable cible qui est (PRIX_DU_LOGEMENT)
et nos varibales clé qui sont dans la liste x
seront séparé en jeu de test et d'entrainment 

l'objectif ici est d'obtenir une bonne base pour l'entrainement 
car il est important de noter que la performance de notre modèle y dépedend 
voila la raison por laquelle on a prit 80% de notre data set pour la dédier
au jeu d'entrainement , contre seulement 20% pour le test 

il est important de noter que cette partie est cruciale pour 
notre projet car c'est d'elle qu'on  saura si notre modele fait du overfitting 
ou soit de l'underfitting 

"""


X_train,X_test,y_train,y_test = train_test_split(x, PRIX_DU_LOGEMENT,test_size=0.2, random_state=0)



```

```python


"""
    lorsqu'on utilise : 
    les algorithme comme :  lasso(L1), Ridge(L2), régression linaire il est important de 
    normaliser pour éviter que le model concidere 
    une features plus que les autre ( c'est comme mettre toutes les variable au meme niveau
    de comparaison )
    la normalisation permet a ridge,lasso, régression linaire de se 
    baser sur l'influences de chaqes variable  d'entre sur la prédiction 
    au lieu de ce concentre uniquement sur leurs valeurs numérque 
"""

scaler = StandardScaler()

X_train_Normaliser = scaler.fit_transform(X_train)
X_test_Normaliser = scaler.transform(X_test)
```
