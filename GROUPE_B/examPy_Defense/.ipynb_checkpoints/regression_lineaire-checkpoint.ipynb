```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler



df = pd.read_csv(r'C:/examPy_Defense/bdd/bddBostonNettoyer_sans_aberrante.csv')

print(df.head())
print(df.info())
print(df.describe())

PRIX_DU_LOGEMENT = df['PRIX_DU_LOGEMENT']
x = df[[
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'CONCENTRATION_NOX',
    'NBR_MOYEN_PIECES',
    'ANCIENNETE_LOGEMENT',
    'DISTENCE_ENTREPRISES',
    'ACCES_AUTOROUTE',
    'TAUX_TAXE_FONCIERE',
    'RATION_ELEVE_ENSEIGNANT',
    'DEMOGRAPHIE_MINORITAIRE',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
]]
```

```python

"""

Notre dataset separé en deux la variable cible qui est (PRIX_DU_LOGEMENT)
et nos varibales clé qui sont dans la liste x
seront séparé en jeu de test et d'entrainment 

l'objectif ici est d'obtenir une bonne base pour l'entrainement 
car il est important de noté que la performance de notre modèle y dépedend 
voila la raison por laquelle on a prit 80% de notre data set pour la dédié 
au jeu d'entrainement , contre seulement 20% pour le test 

il est important de noter que cette partie est cruciale pour 
notre projet car c'est d'elle qu'on  saura si notre modele fait du overfitting 
ou soit de l'underfitting 

"""




X_train,X_test,y_train,y_test = train_test_split(x,PRIX_DU_LOGEMENT,test_size=0.2, random_state=0)

```

```python




"""
    On instancie notre algorithme de regression LinearRegression()
    pour prepare, entrainer et tester notre model

"""

algothme_regression = LinearRegression()

algothme_regression.fit(X_train, y_train)




"""
    on calcul la pente(coef_) pour chaque features c'est a dire commement 
    le prix du logement  vari en fonction de chaque variable

"""
"""

    Noté : pour déterminer la pente la formule est la suivante : 
    Y = ax + b
    ou :
    y = PRIX DU LOGEMENT feature cible
    a =  c'est la pente ou le le prix moyen du logement, c'est la seule 
    valeur qui a été calculée en utilisant toutes les features 
    contenues dans la liste x
    elle répresente la tendance générale de la correlation entre le features(y,x)
    a = coef_

    b = c'est la valeur d'origine , c'est à dire la valeur si toute les clés sont 
    égales à zero(0)

    b = intercept

    formule : 
    a = somme((xi-(x caré).(yi-(y caré)))/somme(xi- (x caé)) le tout au caré
    b =  (y caré) + ax(x caré)


"""

pentes = algothme_regression.coef_


"""
    On affcihe le coefficient de chaque feature sous forme de tableau

"""

print("----------------------On affcihe le coefficient de chaque feature sous forme de tableau---------------------")
df_p = pd.DataFrame(pentes, x.columns, columns=['coeff'])
print(df_p)

"""

    On affcihe la valeur d'origine

"""

print('-------------------------On affcihe la valeur d\'origine---------------------')
print(algothme_regression.intercept_)

```

```python


print("""--ON AFFCIHE LA LISTE DE TOUT LE PRIX DE LOGEMENT POUR LE TEST (y_test) VALEUR REEL --""")
print(f'{y_test}')

print("""--ON AFFCIHE LA LISTE DE TOUT LE PRIX DE LOGEMENT PREDIT PARTANT DU  (x_test)--""")

y_predis = algothme_regression.predict(X_test)
print(f'{y_predis}')

```

```python


"""
    On essaie de voir la correlation qui exite entre les prix des logement predit
    et les vrai prix du logement présant dans le dataset 

"""

plt.scatter(y_test,y_predis)
plt.title('GRAPHE DE PREDICTION DE PRIX DE LOGEMENT  ')
plt.xlabel('PRIX REEL')
plt.ylabel('PRIX DES LOGEMENT PREDIS ')
plt.show()

```

```python


"""

En c'est qui concern le 
MAE : Moyenne absolue d'erreur, c'est à dire on essaie de voir à  combien de dégré
notre model peut se tromper 
posons que 
le y_test = [10,20,30]
le y_predis = [12,18,32]

calcul d'erreur : 
MAE = (y-yi)*(y-yi)
donc : 10-12 = 2
       20-18 = 2
       30-32 = -2
erreur au carée = 2*2 = 4
                  2*2 = 4
                  -2*-2 = 4 
MAE = 12/3 = |4| = 4 


MES = En c'est qui concern le 
MES : Moyenne quadratique d'erreur, c'est à dire on essaie de voir à  combien de dégré 
notre model peut se tromper 
posons que 
le y_test = [10,20,30]
le y_predis = [12,18,32]

calcul d'erreur : 
MES = (y-yi)*(y-yi)
donc : 10-12 = 2
       20-18 = 2
       30-32 = -2
erreur au carée = 2*2 = 4
                  2*2 = 4
                  -2*-2 = 4
MES = 12/3 = 4

"""

"""

 Note :MES et MAE    teste la perforcemence de notre model avec 
    MSE "plus se proche du zeros mieux c'est"
    R2 "plus se proche de 1 mieux c'est"



"""


MAE = mean_absolute_error(y_test,y_predis)
MES = mean_squared_error(y_test,y_predis)
"""R2 Coefficient de détermination , plus se proche de 1 , cela indique notre modèle est bon"""

r2 = r2_score(y_test,y_predis)
RMES = np.sqrt(MES)

print(f'mean_absolute_error {MAE}')
print(f'mean_squared_error {MES}')
print(f'Coefficient de détermination R2 {r2}')
print(f'la racine carre du mean_squared_error {RMES}')

```

```python

"""
    LA COURBE DE RESIDUS(ERREURS)
    ce graphique nous permet de voir la distribution (répartition d'erreurs)
    d'erreurs entre le valeurs prédis et le vrai valeurs observé dans l'univers réel
    erreurs = y(reel) - y(predis)
    notes : 
        une distribution aleatoire : est un bon 
        signe cela nous montre que notre modèle n'est se trompe 
        pas assez (faible taux d'ereurs), les erreurs sont bien repartie

        tendance(ou des motif ) : cela veut dire que 
        notre modèle commet des érreurs 

"""

"""calcul des résidus"""

residus = y_test - y_predis

plt.figure(figsize=(10,6))
sns.residplot(x = y_predis, y = residus,color='g')
plt.title('COURBE DE RESIDUS')
plt.xlabel('Prix des logement prédis')
plt.ylabel('residus')
plt.grid(True)
plt.show()
```
