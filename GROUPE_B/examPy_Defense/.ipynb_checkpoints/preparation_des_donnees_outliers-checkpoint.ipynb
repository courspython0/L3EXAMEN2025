```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,LassoCV
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import IsolationForest





df = pd.read_csv(r'C:/examPy_Defense/bdd/bddBostonNettoyer.csv')


print(df.describe())


```

```python


"""
    Avec notre actuel dataframe nous avons constanté qu'il y a des valeurs abérrantes

    une valeur abérrante est une valeur qui est differente de l'ensemble des valeurs 
    observé dans notre dataframme ou soit une valeur superieure à la moyenne
    des valeurs du dataset 
    on la detecte souvent à l'aide du Boxplot

    les valeurs qui sont superieur au Q3 sont des valeurs aberrante extreme
    les valeurs qui sont inferieur au Q1 sont des valeurs aberrante inferieur

    Notes : 
        Il est important de noter que la presence des Outliers dans notre df 
        reduit la performance et la capaciter à deduire le prix pour notre modele
        de boston 

    de ce fait  de deux  logique sont à considerer  
    detrminer les Outliers en partant de 
        - Qantile  : pour ce faire on calcul le quantile d'odre un et d'odre trois
        - Z-score  :  
    il est important de savoir que ce deux technique son important 
    pour les distribution ou (df) non symetrique , une autre difficulte est 
    qu'on determine les outliers seulement pour un seul feature (univarié)


    la meilleure solution est d'utiliser la classe IsolationForest de la bibliothèque 
    sklearn, avec qu'elle on detecte le outlieurs sur l'ensemble des features de notre df
    en plus IsolationForest travail avec des distribution non symetrique et symetrique
    elle aussi(multivarié)

    logique avec IsolationForest : elle construit plusieur arbdre des décision(sous-ensemble du df)
    et par de la logique ou toute les valeurs des features differente de l'ensemble des 
    valeurs du dataset sont des outliers , et fait la meme chose pour chaque sous-ensemble
"""


"""Ici on utilise une methode manuelle pour détecter les Outliers  """




def detection_Outlier(dataset, colonne):

    Q1 = dataset[colonne].quantile(0.25)
    Q3 = dataset[colonne].quantile(0.75)
    IQ = Q3-Q1
    """On calcule le quuantile d'ordre un"""
    valeur_aberrantes_inferieur = Q1 - 1.5 * IQ
    """On calcule le quuantile d'ordre trois"""
    valeur_aberrantes_superieur = Q3 + 1.5 * IQ

    return valeur_aberrantes_inferieur, valeur_aberrantes_superieur



for colonne in df.columns:
    if df[colonne].dtype in ['int64','float64']:
        
        df[colonne] = np.log(df[colonne] + 1)

        
        valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier(df, colonne)

        df.loc[df[colonne]<valeur_aberrantes_inferieur, colonne] = valeur_aberrantes_inferieur
        df.loc[df[colonne]>valeur_aberrantes_superieur, colonne] = valeur_aberrantes_superieur



```

```python

"""Détection des outliers avec le Zscore"""

def detection_Outlier_zScore(dataset, colonne,limite = 3):
    
    moyen_dataset = dataset[colonne].mean()
    ecart_type = dataset[colonne].std()

    valeur_aberrantes_inferieur = moyen_dataset - limite *ecart_type
    valeur_aberrantes_superieur = moyen_dataset + limite *ecart_type
    return valeur_aberrantes_inferieur, valeur_aberrantes_superieur




for colonne in df.columns:
    if df[colonne].dtype in ['int64','float64']:

        valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier_zScore(df, colonne)

        df.loc[df[colonne]<valeur_aberrantes_inferieur, colonne] = valeur_aberrantes_inferieur
        df.loc[df[colonne]>valeur_aberrantes_superieur, colonne] = valeur_aberrantes_superieur





```

```python




"""Application de IsolationForest pour detecter tout les Outlliers présent dans mon dattaset"""

"""Instanciation """
iso_forest = IsolationForest(contamination = 0.5 ,  random_state=42)

"""Isolations des outliers"""
outlier_valeurs = iso_forest.fit_predict(df)


"""
    la variable  n'est va contenir que deux valeurs (-1 et 1) -1 veut dire qu'il y 
    a belle est bien un outliers dans cette feateure et 1 veut dire le contraire
"""
print(outlier_valeurs)

df['outliers'] = outlier_valeurs
outliers = df[df['outliers']== -1]
print('les valeur aberante détéctés')
print(outliers)


df_final_init_ = df[df['outliers'] == 1]





```

```python


def detection_Outlier_Features(dataset, colonne):
    Q1 = dataset[colonne].quantile(0.25)
    Q3 = dataset[colonne].quantile(0.75)
    IQ = Q3-Q1
    """On calcule le quuantile d'ordre un"""
    valeur_aberrantes_inferieur = Q1 - 1.5 * IQ
    """On calcule le quuantile d'ordre trois"""
    valeur_aberrantes_superieur = Q3 + 1.5 * IQ

    return valeur_aberrantes_inferieur, valeur_aberrantes_superieur



"""
    Après avoir appliquer nos méthode (Le Zscore,la quantile, IsolationForest)
    Nous Constatons toujours des valeurs abérrante au  sein de ces variable
    voilà pourquoi on applique notre fonction (detection_Outlier_Features)
    sur chaque variable
    réfairez-vous de c'est qu'on se dit ci-haut le Zscore et la methode de la 
    quantile d'ordre 1 et 3  peut s'appliquer sur une seule variable tandisque 
    IsolationForest n'est possible sur l'ensemble des données du dataset

"""
```

```python


TAUX_DE_CRIMINALITE = 'TAUX_DE_CRIMINALITE'

valeur_aberrantes_superieur = detection_Outlier_Features(df_final_init_, TAUX_DE_CRIMINALITE)

valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier_Features(df_final_init_, TAUX_DE_CRIMINALITE)

df_final_init_.loc[df_final_init_['TAUX_DE_CRIMINALITE']<valeur_aberrantes_inferieur, TAUX_DE_CRIMINALITE] = valeur_aberrantes_inferieur
df_final_init_.loc[df_final_init_['TAUX_DE_CRIMINALITE']>valeur_aberrantes_superieur, TAUX_DE_CRIMINALITE] = valeur_aberrantes_superieur




PRIX_DU_LOGEMENT = 'PRIX_DU_LOGEMENT'

valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier_Features(df_final_init_, PRIX_DU_LOGEMENT)

df_final_init_.loc[df_final_init_['PRIX_DU_LOGEMENT']<valeur_aberrantes_inferieur, PRIX_DU_LOGEMENT] = valeur_aberrantes_inferieur
df_final_init_.loc[df_final_init_['PRIX_DU_LOGEMENT']>valeur_aberrantes_superieur, PRIX_DU_LOGEMENT] = valeur_aberrantes_superieur





NBR_MOYEN_PIECES = 'NBR_MOYEN_PIECES'

valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier_Features(df_final_init_, NBR_MOYEN_PIECES)

df_final_init_.loc[df_final_init_['NBR_MOYEN_PIECES']<valeur_aberrantes_inferieur, NBR_MOYEN_PIECES] = valeur_aberrantes_inferieur
df_final_init_.loc[df_final_init_['NBR_MOYEN_PIECES']>valeur_aberrantes_superieur, NBR_MOYEN_PIECES] = valeur_aberrantes_superieur



ANCIENNETE_LOGEMENT = 'ANCIENNETE_LOGEMENT'

valeur_aberrantes_inferieur, valeur_aberrantes_superieur = detection_Outlier_Features(df_final_init_, ANCIENNETE_LOGEMENT)

df_final_init_.loc[df_final_init_['ANCIENNETE_LOGEMENT']<valeur_aberrantes_inferieur, ANCIENNETE_LOGEMENT] = valeur_aberrantes_inferieur
df_final_init_.loc[df_final_init_['ANCIENNETE_LOGEMENT']>valeur_aberrantes_superieur, ANCIENNETE_LOGEMENT] = valeur_aberrantes_superieur



```
