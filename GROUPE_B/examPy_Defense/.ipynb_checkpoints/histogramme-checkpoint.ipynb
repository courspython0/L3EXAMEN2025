```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler


df = pd.read_csv(r'C:/examPy_Defense/bdd/bddBostonNettoyer_sans_aberrante.csv')

PRIX_DU_LOGEMENT = df['PRIX_DU_LOGEMENT']
x = df[[
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'CONCENTRATION_NOX',
    'NBR_MOYEN_PIECES',
    'ANCIENNETE_LOGEMENT',
    'DISTENCE_ENTREPRISES',
    'ACCES_AUTOROUTE',
    'TAUX_TAXE_FONCIERE',
    'RATION_ELEVE_ENSEIGNANT',
    'DEMOGRAPHIE_MINORITAIRE',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
]]



""" visualisation des données avec matplotlib Histogramme"""

print(""" visualisation des données avec matplotlib Histogramme""")

"""
    le bins permet de repartir les features composant nos cles (rappelons que ces variable permettent 
    de déterminer le prix d'une maison a boston.
    il regroupes ces variable en termes des groupes (intervalle ).
    Note : pour déterminer la valeurs du bins ont fait : bins = 1 + log(2)N
    ou N = nombres de lignes de notre dataset 
    log(2) = on calcul le logarithme en base 10 de 2
    log2(x) = log10(N)/log10(2) = 8,62
    bins = 1 + 8,62 = 9,62 on arrondit à 10
    géneralement le bin est de 20 , mais toute fois le bins peut-etre calculer
"""
```

```python

"""On visualise la variable PRIX_DU_LOGEMENT"""

plt.hist(PRIX_DU_LOGEMENT, bins = 10 , color = 'skyblue', edgecolor = 'black')

plt.title('Distribution du PRIX_DU_LOGEMENT')
plt.xlabel('PRIX_DU_LOGEMENT')
plt.ylabel('frequence')
plt.show()
```
