```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler


df = pd.read_csv(r'C:/examPy_Defense/bdd/bddBostonNettoyer_sans_aberrante.csv')

df = df.drop('outliers', axis = 1)

PRIX_DU_LOGEMENT = df['PRIX_DU_LOGEMENT']
x = df[[
    'TAUX_DE_CRIMINALITE',
    'zonage_residentiel',
    'PROPORTION_INDUSTRIELLE',
    'CONCENTRATION_NOX',
    'NBR_MOYEN_PIECES',
    'ANCIENNETE_LOGEMENT',
    'DISTENCE_ENTREPRISES',
    'ACCES_AUTOROUTE',
    'TAUX_TAXE_FONCIERE',
    'RATION_ELEVE_ENSEIGNANT',
    'DEMOGRAPHIE_MINORITAIRE',
    'STATUT_SOCIO_ECONOMIQUE_INFERIEUR',
]]
```

```python


""" visualisation des données avec Heatmap seaborn"""

print(""" visualisation des données avec Heatmap seaborn""")





"""
    
    On confirme la correlation entre variables (voir la matrice de poissons )
    Note: par défaut nos varibales présent dans le dataset ne 
    sont pas en rélation, voilà pourquoi la matrice heatmap , necessite qu'il ait correlation 
    entre les varibale avant sans mise en place 
    sans laquelle il y aura erreur


"""
matrice_de_correlation = df.corr()

```

```python

"""
    On creer le heatmap et on lui donne la taille de 12 
    sur la longuere et 10 sur la largeur
    
"""
"""

   annot=True permet d'afficher les valeurs de correlations sur le graphique
   cmap='coolwarm' definit la pallete des couleurs 
   fmt='.2f' format le nombre avec 2 decimals

"""

plt.figure(figsize=(12,10))
sns.heatmap(matrice_de_correlation, annot=True,cmap='coolwarm',fmt='.2f')
plt.title('HEATMAP')
plt.show()


"""
    palette de couleurs
    rouge : la correlation est positif
    bleu : la corelation est négatif
    palle : pas de corelation



"""
```
